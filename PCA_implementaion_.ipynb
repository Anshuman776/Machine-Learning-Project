```python
import numpy as np
import pandas as pd
data = np.array([[3, 7],
                 [-4, -6],
                 [1, -1],
                 [7, 8],
                 [-4, -1],
                 [-3, -7]])
dataframe = pd.DataFrame(data, columns=['x1', 'x2'])
```

```python
dataframe.describe()
```

standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
datascaled = scaler.fit_transform(dataframe)
```

```python
dataframe.describe()
```

```python
datascaled_df = pd.DataFrame(datascaled, columns=['x1', 'x2'])
display(datascaled_df.describe())
```

```python
datascaled_df.describe()
```

```python
c1 = dataframe.x1
c2 = dataframe.x2
np.cov(c1, c2)
```

```python
#  Approach covarience
c1 = datascaled_df.x1
c2 = datascaled_df.x2
np.cov(c1, c2)
```

```python
# Approach 2
covarienceMatrix = datascaled_df.T @ datascaled_df/5
covarienceMatrix
```

```python
#  Approach 3
np.sum(c1*c2)/5
```

Eigen values and Eigen Vectors

```python
eigenValues,eigenVectors = np.linalg.eig(covarienceMatrix)
```

calculating the new features or principle components

```python
## PC1 - Contains the maximum information of the original two features that you have
PC1 = datascaled_df @ eigenVectors[:, 0]
## PC2
PC2 = datascaled_df @ eigenVectors[:, 1]
```
